````python
import requests
````

Prompt für ChatGPT: ```Can show me the API request I need to write, to get all the names, capitals, population of all the countries in the REST Countries API? Can you show me using Python?```

````python
# Code from CHATGPT. Delivered with 

url = "https://restcountries.com/v2/all?fields=name,capital,population"

response = requests.get(url)

if response.status_code == 200:
    countries = response.json()
    for country in countries:
        name = country.get("name", "N/A")
        capital = country.get("capital", "N/A")
        population = country.get("population", "N/A")
        print(f"Name: {name}, Capital: {capital}, Population: {population}")
else:
    print(f"Error {response.status_code}: Unable to fetch data.")
````

Prompt für ChatGPT: ```I don't want to print the result, but would prefer it in a Pandas dataframe I can save off. Can you help me?```

````python
import requests
import pandas as pd

url = "https://restcountries.com/v2/all?fields=name,capital,population,languages"

response = requests.get(url)

if response.status_code == 200:
    countries = response.json()
    # Create a DataFrame from the JSON data
    df = pd.DataFrame(countries)
    
    # Save the DataFrame to a CSV file
    df.to_csv('countries_data.csv', index=False)
else:
    print(f"Error {response.status_code}: Unable to fetch data.")
````
